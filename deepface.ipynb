```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
img1= cv2.imread("/content/ayon-saddar-225x300.jpeg.crdownload")
plt.imshow(img1)
plt.show()
```

```python
#plotting the image
#Flipping the color from BGR TO RGB 
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
obj=DeepFace.analyze(img1,actions=['age','gender','race','emotion']) #We are using DeepFace library 
print(obj)
```

```python
##print result
print("a)Your age is :",obj["age"], 
      "\nb)You are : ",obj["dominant_race"],
      "\nc)Current emotion is :",obj["dominant_emotion"],
      "\nd)Your Gender :",obj["gender"])
```
