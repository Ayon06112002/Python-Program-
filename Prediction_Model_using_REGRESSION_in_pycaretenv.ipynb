```python
from pycaret.datasets import get_data
```

```python
dataSets=get_data('index')
```

```python
ins_1=get_data("insurance")
```

```python
from pycaret.regression import *
s_model= setup(data=ins_1, target ='charges', silent = True)
```

```python
models()
```

```python
lasso= create_model('lasso')
```

```python
svm=create_model('svm')
```

```python
best_model=compare_models(fold=3) #we can change the number of folds. By default it is ten.
```

```python
#comparing two models 
best_model=compare_models(include=['svm','lasso'])
```
